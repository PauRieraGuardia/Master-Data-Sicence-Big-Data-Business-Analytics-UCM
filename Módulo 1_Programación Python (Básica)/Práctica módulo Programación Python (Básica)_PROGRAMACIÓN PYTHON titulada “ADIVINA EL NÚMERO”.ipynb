```python
import random as R
import pandas as pd

resultados = 'C:/Users/pauri/OneDrive/Escritorio/Tarea Módulo 1/resultados_juego.xlsx' # IMPORTANTE! Debes indicar la ruta donde quieres que se guarde el archivo xlsx

def menu():
    print('1. Partida modo solitario')
    print('2. Partida 2 jugadores')
    print('3. Estadística')
    print('4. Salir ')


def submenu():
    print('1. Fácil (20 intentos)')
    print('2. Medio (12 intentos)')
    print('3. Difícil (5 intentos)')
    print('4. Salir al menu principal')

def guardar_resultado(nombre, dificultad, resultado, intentos):
    try:
        df = pd.read_excel(resultados)
    except FileNotFoundError:
        df = pd.DataFrame(columns=['Nombre','Dificultad','Resultado','Intentos'])

    nueva_fila = pd.DataFrame([{
        'Nombre': nombre,
        'Dificultad': dificultad,
        'Resultado': resultado,
        'Intentos': intentos
      }])

    df = pd.concat([df,nueva_fila],ignore_index=True)
    df.to_excel(resultados,index=False)

def estadisticas():
    try:
        df = pd.read_excel(resultados)
        print(df)
    except FileNotFoundError:
        print("Aún no hay estadísticas disponibles.")

def un_jugador():
    nombre = input('Introduce tu nombre: ')
    while True:
        submenu()
        dificultad = int(input('Introduce un nivel de dificultad: '))

        if dificultad == 1:
            numero1 = R.randint(1,1000)
            numero2 = int(input('Introduce un número: ',))
            contador = 1
            while numero1 != numero2:
                if contador >= 20:
                    print('Has perdido')
                    guardar_resultado(nombre,'Fácil','Derrota',20)
                    return
                if numero1 > numero2:
                    print('Has fallado, el número es mayor')
                else:
                    print('Has fallado, el número es menor')

                numero2 = int(input('Prueba con otro número: ',))
                contador = contador + 1
            else:
                print('¡Has acertado!')
                guardar_resultado(nombre,'Fácil','Victoria',contador)
                break

        if dificultad == 2:
            numero1 = R.randint(1,1000)
            numero2 = int(input('Introduce un número: ',))
            contador = 1
            while numero1 != numero2:
                if contador >=12:
                    print('Has perdido')
                    guardar_resultado(nombre,'Medio','Derrota',12)
                    return
                if numero1 > numero2:
                    print('Has fallado, el número es mayor')
                else:
                    print('Has fallado, el número es menor')

                numero2 = int(input('Prueba con otro número: ',))
                contador = contador + 1
            else:
                print('¡Has acertado!')
                guardar_resultado(nombre,'Medio','Victoria',contador)
                break

        if dificultad == 3:
            numero1 = R.randint(1,1000)
            numero2 = int(input('Introduce un número: ',))
            contador = 1
            while numero1 != numero2:
                if contador >= 5:
                    print('Has perdido')
                    guardar_resultado(nombre,'Difícil','Derrota',5)
                    return
                if numero1 > numero2:
                    print('Has fallado, el número es mayor')
                else:
                    print('Has fallado, el número es menor')

                numero2 = int(input('Prueba con otro número: ',))
                contador = contador + 1
            else:
                print('¡Has acertado!')
                guardar_resultado(nombre,'Difícil','Victoria',contador)
                break

        if dificultad == 4:
            break

        if dificultad <1 or dificultad >4:
            print('Opción Incorrecta')

def dos_jugadores():
    nombre1 = input('Jugador 1, introduce tu nombre: ')
    nombre2 = input('Jugador 2, introduce tu nombre: ')
    while True:
        submenu()
        dificultad_2 = int(input('Introduce un nivel de dificultad: '))
        if dificultad_2 == 1:
            numero_jugador_1 = int(input('Jugador 1, piensa un número: ',))
            numero_jugador_2 = int(input('Jugador 2, adivina el número: ',))
            contador_2 = 1
            while numero_jugador_1 != numero_jugador_2:
                if  contador_2 >= 20:
                    print('Has perdido')
                    guardar_resultado(nombre2,'Fácil','Derrota',20)
                    return
                if numero_jugador_1 > numero_jugador_2:
                    print('Has fallado, el número es mayor')
                else:
                    print('Has fallado, el número es menor')
                numero_jugador_2 = int(input('Prueba con otro numero: ',))
                contador_2 = contador_2 + 1
            else:
                print('¡Has acertado!')
                guardar_resultado(nombre2,'Fácil','Victoria',contador_2)
                break

        elif dificultad_2 == 2:
            numero_jugador_1 = int(input('Jugador 1, piensa un número: ',))
            numero_jugador_2 = int(input('Jugador 2, adivina el número: ',))
            contador_2 = 1
            while numero_jugador_1 != numero_jugador_2:
                if  contador_2 >= 12:
                    print('Has perdido')
                    guardar_resultado(nombre2,'Medio','Derrota',12)
                    return
                if numero_jugador_1 > numero_jugador_2:
                    print('Has fallado, el número es mayor')
                else:
                    print('Has fallado, el número es menor')
                numero_jugador_2 = int(input('Prueba con otro numero: ',))
                contador_2 = contador_2 + 1
            else:
                print('¡Has acertado!')
                guardar_resultado(nombre2,'Medio','Victoria',contador_2)
                break

        elif dificultad_2 == 3:
            numero_jugador_1 = int(input('Jugador 1, piensa un número: ',))
            numero_jugador_2 = int(input('Jugador 2, adivina el número: ',))
            contador_2 = 1
            while numero_jugador_1 != numero_jugador_2:
                if  contador_2 >= 5:
                    print('Has perdido')
                    guardar_resultado(nombre2,'Difícil','Derrota',5)
                    return
                if numero_jugador_1 > numero_jugador_2:
                    print('Has fallado, el número es mayor')
                else:
                    print('Has fallado, el número es menor')
                numero_jugador_2 = int(input('Prueba con otro numero: ',))
                contador_2 = contador_2 + 1
            else:
                print('¡Has acertado!')
                guardar_resultado(nombre2,'Difícil','Victoria',contador_2)
                break

        elif dificultad_2 == 4:

            break

        elif dificultad_2 < 1 or dificultad_2 > 4:
            print('Opción Incorrecta')

while True:
    menu()
    modo_juego = int(input('Elige un modo de juego: ',))

    if modo_juego == 1:
        un_jugador()

    if modo_juego == 2:
        dos_jugadores()

    if modo_juego == 3:
        estadisticas()

    if modo_juego == 4:
        break

    if modo_juego < 1 or modo_juego > 4:
        print('Opción Incorrecta')
```
