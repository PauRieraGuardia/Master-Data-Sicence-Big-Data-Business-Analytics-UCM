### Mapas con leafet/Folium

Librería que te permite crear mapas interactivos desde Python y los renderiza usando Leaflet por debajo. Ideal para cuadernos Jupyter y para exportar a HTML

```python
import folium # Libreria folium
```

```python
40.43854862426588, -3.7014533047534233 # Coordenadas de la facultad
```

```python
coordenadas_ucm_docu = [40.43854862426588, -3.7014533047534233] # Guarda las coodenadas en una variable llamada coordenadas_ucm_docu
```

```python
m = folium.Map(location = coordenadas_ucm_docu)  # Aplica folium para dibuajar el mapa y lo guarda en una variable m
```

```python
m
```

```python
m = folium.Map(location=coordenadas_ucm_docu, zoom_start = 16) # zoom_start para aplicar zoom de inicio
m
```

```python
lamuno = [43.56155103751697, -6.20080612074845] # Coordenadas Lamuño municipio
```

```python
m = folium.Map(location=lamuno, zoom_start=14)
m
```

```python
folium.TileLayer('Carto DB Positron').add_to(m) # cambiamos el estilo del gráfico
```

```python
m
```

```python
folium.TileLayer('Stamen Terrain',
                 name = 'Stamen Toner',
                 attr = 'Map tiltes by Stamen Design, under CC BY 3.0. Data by OpenStreetMap, undex ODbl').add_to(m)
```

```python
m
```

```python
m = folium.Map(location = coordenadas_ucm_docu,
               zoom_start = 14,
               titles = 'Carto DB Positron')

tooltip = 'UCM docu'

folium.Marker(coordenadas_ucm_docu, # para poner una chincheta en el mapa de las coordenadas
              popup = 'Máster DS',  # texto que quiero que salga cuando clico la chincheta
              tooltip= tooltip).add_to(m) # Texto que aparece cuando paso por encima
```

```python
m
```

```python
# Volvemos al mapa original
map = folium.Map(location=coordenadas_ucm_docu, zoom_start=14)
map
```

```python
tooltip = 'Haz click para ver la dirección'

folium.Marker(
    [40.43847105761711, -3.7014854890675513], # coordenadas facultad turismo ucm
    popup="<a href='https://www.ucm.es/estudios/grado-turismo' target = '_blank'>UCM Turismo</a>", # añadimos link al popup de la facultad de turismo de la UCM
    tooltip=tooltip, # tooltip
     icon=folium.Icon(color="red", icon="info-sign"), # modificamos el icono (Marker) y le ponemos el icono de información
).add_to(map) # lo añadimos al mapa
folium.Marker(
    [40.439889422283926, -3.711783660232115], # coordenadas facultad documentación
    popup="<a href='https://documentacion.ucm.es/' target = '_blank'>UCM Documentación</a>",  # Link facultad documentación
    tooltip=tooltip, # tooltip
    icon=folium.Icon(color="blue"),# Icono a blue
).add_to(map)

map
```
