## 1. Datos veh√≠culos a la venta 2016 en UK

Proposito:

* facilitar la lectura del .csv que contiene los datos de este ejemplo

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import plotnine
from plotnine import *
```

### Carga de datos

```python
df = pd.read_pickle('datos_vehiculos_2016_UK.pkl')
```

```python
df.shape
```

```python
df.describe(include='all')
```

```python
count_classes = pd.value_counts(df['Tipo'],sort=True)
count_classes.plot(kind='bar', rot=0)
plt.title('Tipos de veh√≠culos UK 2016')
plt.xlabel('Tipo')
plt.ylabel('Frecuancia')
```

```python
plt.scatter(df['EngineCapacity'],
            df['MetricCombined'])

plt.xlabel('Tama√±o motor (cm3)')
```

### 2. Visualizaciones con ggplot 

Adem√°s del contenido recogido en la docu. te√≥rica, es aconsejable segur este excelente (y pr√°ctico) tutorial:

https://realpython.com/ggplot-python 

```python
( ggplot(data=df) 
 + aes(x = 'EngineCapacity',
     y = 'MetricCombined')
) # Nos nos dara una forma puesto que no le hemos dado una geometria 
```

```python
( ggplot(data=df) 
 + aes(x = 'EngineCapacity',
     y = 'MetricCombined')
+ geom_point() # puntos
)
```

```python
( ggplot(data=df) 
 + aes(x = 'EngineCapacity',
     y = 'MetricCombined')
+ geom_smooth() # linea de tendencia
)
```

```python
( ggplot(data=df) 
 + aes(x = 'EngineCapacity',
     y = 'MetricCombined')
+ geom_point() # grafico de puntos
+ geom_smooth() # linea de tendencia
)
```

```python
( ggplot(data=df) 
 + aes(x = 'EngineCapacity',
     y = 'MetricCombined',
     color = 'Tipo') 
+ geom_point(alpha=0.1) 
+ geom_smooth(method='lm') # Muestra modelo lineal

)
```

```python
( ggplot(data=df) 
 + aes(x = 'EngineCapacity',
     y = 'MetricCombined',
     color = 'Tipo') 
+ geom_point(alpha=0.1) 
+ geom_smooth(method='lowess') # muestra variaciones no lineales del todo
)
```

```python
g = ( ggplot(data=df) 
 + aes(x = 'EngineCapacity',
     y = 'MetricCombined',
     color = 'CO2gkm') 
+ geom_point(alpha=0.1) 
+ geom_smooth(method='lm') # Muestra modelo lineal
+ geom_smooth(method='lowess')
)
```

```python
g = g + theme(figure_size=(12,12)) # damos un tama√±o a la figura redefiniendo g
```

```python
( ggplot(df) 
 + aes(x = 'EngineCapacity',
     y = 'MetricCombined',
     color = 'Tipo') 
+ geom_smooth(method='lowess')
+ geom_point()
+ facet_grid(cols='Tipo') # Aplciamos una columna de gr√°fico para cada uno
) 
```

```python
( ggplot(df) 
 + aes(x = 'EngineCapacity',
     y = 'MetricCombined',
     color = 'Tipo') 
+ geom_smooth(method='lowess')
+ geom_point()
+ facet_grid(rows='Tipo') # Aplciamos una fila de gr√°fico para cada uno
) 
```

```python
( ggplot(df) 
 + aes(x = 'Tipo',
     y = 'EmissionsNOxmgkm') 
+ geom_boxplot() # Grafico boxplot
) 
```
