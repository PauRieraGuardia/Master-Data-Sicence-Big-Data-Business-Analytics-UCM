#### 2. Realizar una petición GET a la API

Utilicen requests para obtener la información sobre las obras de artes. Extraigna al menos los siguientes campos (La información de los campos los podéis encontrar en: https://api.artic.edu/docs/#conventions)

```python
import requests
import pandas as pd
response = requests.get('https://api.artic.edu/api/v1/artworks').json() # accedemos a la API, que es u  json
```

```python
response.keys() # utilizamos keys() para obtener las claves de un diccionario
```

```python
for clave in response.keys(): # Hacer un print de las keys del diccionario
    print(clave)
```

```python
response.items() # Para acceder a toda la información del diccionario
```

```python
response.values() # Para acceder a los valores del diccionario
```

```python
df = pd.DataFrame(response['data']) # Creamos un dataframe a través de la clave-valor data
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

#### 3. Análisis de los Datos con Pandas

Una vez obtendio el DataFrame, respondan las siguientes preghntas usando pandas:

¿Cuántas obras en total se han obetenido?

¿Cuántas obras tienen información sobre su artista?

¿Cuáles son los 5 artistas con más obras en la base de datos obtenida?

```python
df.count() # Obras en total 
```

```python
# Cuantas obras tienen información sobre su artista
# Ambos métodos son lo mismo, 
# En el primero le decimos que nos saque los que no son nulos 
# En el segundo le indicamos que nos saque la negación de los que son nulos
print(len(df[df['artist_id'].notna()]))
print(len(df[~df['artist_id'].isna()])) # negacion de is null
```

```python
# 5 artistas con mas obras
df.groupby('artist_title', as_index=False).agg(count=('id','count')).sort_values(by='count', ascending=False)[:5]
```
