### 3. Generar los dataframes. "aplanando los diccionarios" y generando columnas nuevas para cada cliente

Dataframe orders

```python
df_orders = df_customers[['CustomerEmail','orders']]
```

```python
# Otro método para extraer las columnas, pasarlo a diccionario solo como elemento
df_orders= df_orders.explode('orders')
```

```python
df_orders['orders'][0] # accedemos al primer elemento y vemos que solo hay dos filas, pues es diccioanrio, el indice es incorrecto
```

```python
df_orders.reset_index(drop=True,inplace=True)
```

```python
df_orders['orders'][0] # Indice corregido
```

```python
def extract_json(dict_, value):
    try:
        return dict_.get(value)
    except:
        return None
```

```python
df_orders['orderId'] = df_orders['orders'].apply( lambda x:extract_json(x,'orderId'))
df_orders['orderAmount'] = df_orders['orders'].apply( lambda x:extract_json(x,'orderAmount'))
df_orders['orderState'] = df_orders['orders'].apply( lambda x:extract_json(x,'orderState'))
df_orders['orderShippingAddress'] = df_orders['orders'].apply( lambda x:extract_json(x,'orderShippingAddress'))
df_orders.drop('orders',axis=1,inplace=True)
df_orders
```

Dataframe payment methods

```python
df_paymentMethods = df_customers[['CustomerEmail','paymentMethods']]
```

```python
df_paymentMethods= df_paymentMethods.explode('paymentMethods')
```

```python
df_paymentMethods.reset_index(drop=True,inplace=True)
```

```python
df_paymentMethods['paymentMethods'][0]
```

```python
df_paymentMethods['paymentMethodId'] = df_paymentMethods['paymentMethods'].apply( lambda x:extract_json(x,'paymentMethodId'))
df_paymentMethods['paymentMethodRegistrationFailure'] = df_paymentMethods['paymentMethods'].apply( lambda x:extract_json(x,'paymentMethodRegistrationFailure'))
df_paymentMethods['paymentMethodType'] = df_paymentMethods['paymentMethods'].apply( lambda x:extract_json(x,'paymentMethodType'))
df_paymentMethods['paymentMethodProvider'] = df_paymentMethods['paymentMethods'].apply( lambda x:extract_json(x,'paymentMethodProvider'))
df_paymentMethods['paymentMethodIssuer'] = df_paymentMethods['paymentMethods'].apply( lambda x:extract_json(x,'paymentMethodIssuer'))
df_paymentMethods.drop('paymentMethods',axis=1,inplace=True)
df_paymentMethods
```

Dataframe transactions

```python
df_transactions = df_customers[['CustomerEmail','transactions']]
```

```python
df_transactions= df_transactions.explode('transactions')
```

```python
df_transactions.reset_index(drop=True,inplace=True)
df_transactions['transactions'][0]
```

```python
df_transactions['transactionId'] = df_transactions['transactions'].apply( lambda x:extract_json(x,'transactionId'))
df_transactions['orderId'] = df_transactions['transactions'].apply( lambda x:extract_json(x,'orderId'))
df_transactions['paymentMethodId'] = df_transactions['transactions'].apply( lambda x:extract_json(x,'paymentMethodId'))
df_transactions['transactionAmount'] = df_transactions['transactions'].apply( lambda x:extract_json(x,'transactionAmount'))
df_transactions['transactionFailed'] = df_transactions['transactions'].apply( lambda x:extract_json(x,'transactionFailed'))
df_transactions.drop('transactions',axis=1,inplace=True)
df_transactions
```
