5. Mergear los dataframes user_usage y user_device (columnas = ['use_id', 'platform', 'device']), por clave 'use_id' y utilizando los diferentes métodos de how (inner, outer, left, right)
Comentar los resultados

```python
df_merged_0 = pd.merge(user_usage,user_device[['use_id', 'platform', 'device']], on='use_id',how='inner')
df_merged_0
```

```python
df_merged_1 = pd.merge(user_usage,user_device[['use_id', 'platform', 'device']], on='use_id',how='outer')
df_merged_1
```

```python
df_merged_2 = pd.merge(user_usage,user_device[['use_id', 'platform', 'device']], on='use_id',how='left')
df_merged_2
```

```python
df_merged_3 = pd.merge(user_usage,user_device[['use_id', 'platform', 'device']], on='use_id',how='right')
df_merged_3
```

6. Utilizar la función groupby en algunos de los dataframes resultantes anteriores, agrupando por la variable "platform" y haciendo una suma del resto de columnas
