# Linear Regression (Surpevised)

```python
# Importamos las librerias necesarias para crear un dataset de manera aleatoria
import matplotlib.pyplot as plt
import numpy as np

# generar datos de entrenamiento
rng = np.random.RandomState(42) #función para generar números random
x = 10 * rng.rand(50) # 50 números aleatorios entre 0 y 1 guardados en la variable x
y = 2 * x - 1 + rng.randn(50) # Multiplica por 2 y resta 1 y le suma un valor random a cada uno de los 5 valores de x

print('x')
print(x)
print()
print('y')
print(y)
print()

plt.scatter(x, y); #Dibujamos los datos de x e y
```

# Unsupervised Iris dimensionality

```python
from sklearn.decomposition import PCA # Importamos el modelo PCA
import seaborn as sns
import matplotlib.pyplot as plt

model = PCA(n_components=2) # Definimos el modelo PCA con 2 n_componentes
model.fit(X_iris) # Lo aplicamos al dataset iris
X_2D = model.transform(X_iris) # Transformamos en 2 dimensiones
X_2D
```

```python
X_2D.shape
```

```python
iris['PCA1'] = X_2D[:, 0] 
iris['PCA2'] = X_2D[:, 1] 
sns.lmplot(x="PCA1", y="PCA2", hue='species', data=iris, fit_reg=False);
plt.show()
```

# Gaussian Mixture Models

```python
from sklearn import mixture
model = mixture.GaussianMixture(n_components=3, covariance_type='full') # Creaun modelo con 3 clusters, suponiendo varianza completa entre variables
model.fit(X_iris) # lo aplicanos al data set iris
y_gmm = model.predict(X_iris) # predecimos en eld ataset iris
iris['cluster'] = y_gmm # añade al dataset iris una nueva columna llamada clusetrs con los valores predichos
sns.lmplot(x="PCA1", y="PCA2", data=iris, hue='species', col='cluster', fit_reg=False); # crea un gráfico de dispersión por cada cluster
plt.show()
```

# Model Validation

```python
from sklearn.datasets import load_iris 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.metrics import accuracy_score 

iris = load_iris() 
X = iris.data
y = iris.target

model = KNeighborsClassifier(n_neighbors=1)
model.fit(X, y) 
y_model = model.predict(X)

accuracy_score(y, y_model)
```

```python
from sklearn.model_selection import train_test_split 
X1, X2, y1, y2 = train_test_split(X, y, random_state=0, train_size=0.5) 

model.fit(X1, y1) 
y2_model = model.predict(X2) 
accuracy_score(y2, y2_model)
```

```python
y2_model = model.fit(X1, y1).predict(X2) 
y1_model = model.fit(X2, y2).predict(X1) 
accuracy_score(y1, y1_model), accuracy_score(y2, y2_model)
```

```python
from sklearn.model_selection import cross_val_score 
scores = cross_val_score(model, X, y, cv=5)
scores
```

```python
scores.mean()
```

```python
from sklearn.model_selection import LeaveOneOut 
scores = cross_val_score(model, X, y, cv=LeaveOneOut()) 
scores
```

```python
scores.mean()
```
