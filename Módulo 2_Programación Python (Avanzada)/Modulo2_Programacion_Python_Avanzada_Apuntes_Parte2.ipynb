#### Dataframes

Un DataFrame es una estructura de datos en Pandas que permite almacenar y manipular datos en forma de tabla con filas y columnas. Puede contener distintos tipos de datos, como números enteros o flotantes, cadenas de texto, fechas y hora, y otros tipos de datos. Cada columna de un DataFrame puede tener un tipo de datos diferente, pero todas las filas en una columna tienen que tener el mismo tipo de datos.

##### Filtrado

Los datos se pueden filtar proporcionando alguna expresión booleana en el DataFrame.

```python
despues_85 = titles[titles['year'] > 1985] # Estamos filtrando los datos del dataframe titles de la columna year que sean mayores a 1985
despues_85 # Nos devuelve otro DataFrame
```

Cuando pasamos los resultados booleanos al DataFrame, Pandas mostrará todos los resultados que sean True. Además, podemos usar "&" (and) "|" (or) para unir dos condiciones.

```python
peliculas_90 = titles[(titles['year']>=1990) & (titles['year']<2000)] # IMPORTANTE!! El and y el or de pandas son & y |
peliculas_90
```

```python
# Estamos seleccionando todos aquellos datos que no son del 1990, el operador != es not
titles[titles['year'] != 1990]
```

#### Agrupación de datos

Los datos se pueden agrupar por columnas-cabeceras. Además se pueden definir formatos personalziados para agrupar los diversos elementos del DataFrame.

##### Agrupación por nombre de columna

El método groupby() devuelve un objeto, al que luego le concatenamos el método size() para contar el número total de filas en cada año. Es el mismo que el comando count_values()

```python
casts_agrupado = casts.groupby(['year']).size()  # .size() es necesario para que nos contabilice los años
casts_agrupado.plot()
```

```python
casts_aaron_abrams = casts[casts["name"] == "Aaron Abrams"]
casts_aaron_abrams.groupby(['year']).size()
```

```python
# Si queremos ver la lista de películas con los titulos agrupados por años debemos pasar las columnas year y title
casts_aaron_abrams.groupby(['year','title']).size()
```

```python
# Queremos agrupar los artículos por año y ver la calificación máxima en esos años

casts_aaron_abrams.groupby(['year']).n.max() #ponemos n pork es la variable n (puntuación) que queremos analizar
```

```python
# Queremos agrupar los artículos por año y ver la calificación mínima en esos años

casts_aaron_abrams.groupby(['year']).n.min()
```

```python
# Queremos agrupar los artículos por año y ver la calificación media en esos años

casts_aaron_abrams.groupby(['year']).n.mean()
```

##### Agrupación por campo custom

#### Stack/Unstack

stack() y unstack() son métodos en Pandas que se utilizan para cambiar la estrucutra de un DataFrame. Stack() convierte columnas en filas, mientras que unstack() convierte filas en columnas. Estos métodos son útiles para transformar los datos en un DataFrame de una forma que sea más fácil de analizar y visualizas.

```python
# Creamos un DataFrame de dos productos y 4 tiendas distintas
df_ventas = pd.DataFrame([
        {'Tienda 1': 100, 'Tienda 2': 150, 'Tienda 3':200, 'Tienda 4': 250},
        {'Tienda 1': 50, 'Tienda 2': 100, 'Tienda 3':150, 'Tienda 4': 200}
])

df_ventas.index = ['Producto 1','Producto 2']
df_ventas
```

```python
# Si quisieramos analizar las ventas por tienda en lugar de por producto, podriams utilizar el método stack() para convertir las columnas en filas

ventas_por_tienda = df_ventas.stack()

ventas_por_tienda
```

```python
# Si quisiéramos analizar las ventas por producto en lugar de por tienda, podríamos utilizar el método unstack() para convertir las finasl en columnas
ventas_por_producto = df_ventas.unstack()
ventas_por_producto
```

```python
# Los datos se agrupan por década y tipo, es decir, actor y actriz

casts.groupby([casts['year']//10*10,'type']).size()
```

Vamos a generar un gráfico con el número total de catores y actrices en cada década y compararemos.

La solución para ello es utilizar unstack() y stack() para comprender mejor los gráficos.

```python
casts_decadas = casts.groupby(['type',casts['year']//10*10]).size()
casts_decadas
```

```python
# Metemos el año como fila y las columnas como filas
casts_decadas.unstack()
```

```python
# Generamos el gráfico
casts_decadas.unstack().plot(kind='bar')
```

```python
# Si queremos representar los datos uno al lado del otro, debemos usar la opcíon usntack(0), por defecto se usa usntack(-1)

casts_decadas.unstack(0)
```

```python
# Hemos cambiado el tipo como columna y el año como fila
casts_decadas.unstack(0).plot(kind='bar')
```

#### Indexación de los datos

En esta sección vamos a analizar un poco más en detalle la importancia de la indexación, que es muy importante en Pandas. Se utiliza para organizar los datos y proporcionarnos un acceso más rápido. En esta sección vamos a comparar los tiempos de acceso a los datos con y sin idnexación y veremos la diferencia.

```python
casts = pd.read_csv('/content/drive/MyDrive/Universidad /4) MÁSTER EN DATA SCIENCE, BIG DATA & BUSINESS ANALYTICS (UCM)/MÓDULO 2: PROGRAMACIÓN EN PYTHON AVANZADA/datasets/cast.csv', encoding = 'ISO-8859-1')
casts
```

```python
# Si usamos el comando %%timeit ejecutará el código Python que viene a continuación varias veces y mostrará el tiempo promedio.
%%timeit -r 5 -n 1000

# -r = para indicar el número de veces que se va a ejecutra el código en cada iteración del loop
# -n el número de iteraciones que queremos ejecutar

casts[casts['title'] == 'Macbeth']
```

```python
# Para hacer un filtro como al anterior, ahora ya que teneos los datos indexados, debemos usar loc[]
%%timeit -r 5 -n 1000

casts_indexados.loc['Macbeth']
```
