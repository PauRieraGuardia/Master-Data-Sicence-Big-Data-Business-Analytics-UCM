### <b>1. Lectura, escritura y visualización de imágenes</b>

#### <b>Lectura de imágenes</b>

OpenCV permite leer imágenes en varios formatos. El método más común es mediante ``2.imread()``.
Esta función carga una imagen desde un archivo y la devuelve como un arreglo de NumPy.
Si la imagen no puede leerse (por falta del archivo, permisos incorrectos o formato no compatible o inválido), la función devuelve una matriz vacía.

* Sintaxis: ``cv2.imread(path, flag)``
* Parámetros:
    * ``cv2.IMREAD_COLOR o 1``: carga una imagen a color. Cualquier transparencia se ignora (modo predeterminado).

    * ``cv2.IMREAD_GRAYSCALE o 0:`` carga la imagen en modo de escala de grises.

    * ``cv2.IMREAD_UNCHANGED o -1:`` carga la imagen tal como está, incluyendo el canal alfa (transparencia).

En OpenCV, cuando cargas una imagen usando cv2.imread(), la imagen se lee por defecto en formato BGR (Blue, Green, Red).
Esto difiere del formato RGB (Red, Green, Blue), que es el más común en muchas otras bibliotecas y aplicaciones de procesamiento de imágenes.

La diferencia entre estos dos formatos es el orden de los canales de color:

* <b>BGR: </b>el canal azul va primero, seguido del verde y luego el rojo.

* <b>RGB:</b> el canal rojo va primero, seguido del verde y luego el azul.

```python
# Conversión de una imagen BGR a RGB:
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)
```

````python
import cv2
import urllib
import matplotlib.pyplot as plt

# Download an image
url = 'http://i.guim.co.uk/img/media/307c4c6b789dcde24a6c1a468452051fd0bd92c4/0_293_1067_640/master/1067.jpg?width=1200&height=900&quality=85&auto=format&fit=crop&s=52fe6f9a377482b700d4843624abf3d1'
urllib.request.urlretrieve(url, "sample.png")


# Path to the image
image_path = "sample.png"

# Read the image in color mode
image_color = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Read the image in grayscale mode
image_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Transform to RGB
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)
````

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns

# Plot original BGR image
axes[0].imshow(image_color)  # Convert BGR to RGB for correct color representation in matplotlib
axes[0].set_title("BGR Image")
axes[0].axis('off')  # Hide axis

# Plot grayscale image
axes[1].imshow(image_gray, cmap='gray')  # Grayscale image
axes[1].set_title("Grayscale Image")
axes[1].axis('off')

# Plot RGB image
axes[2].imshow(image_rgb)  # RGB image
axes[2].set_title("RGB Image")
axes[2].axis('off')

plt.show()
````

#### <b> Escritura de imágenes</b>

Guardar una imagen en OpenCV se realiza mediante la función cv2.imwrite().
Esta función guarda una imagen en un archivo especificado, y el formato de la imagen se determina automáticamente según la extensión del nombre del archivo.

* Sintaxis: ``cv2.imwrite(filename, img, params)``

* Parámetros:
    * <b>filename:</b> cadena que representa el nombre del archivo.
    * <b>img:</b> imagen que se desea guardar.
    * <b>params: </b> parámetros opcionales (por ejemplo, nivel de compresión en archivos JPEG o PNG).


 ### 2. <b>Manipulación básica de imágenes con OpenCV</b>

En esta sección, exploraremos técnicas básicas de manipulación de imágenes como redimensionar, rotar y recortar imágenes utilizando OpenCV.
Estas operaciones son fundamentales en muchas tareas de visión por computador.

#### <b>Redimensionamiento de imágenes</b>

El redimensionamiento es una de las operaciones más comunes en el procesamiento de imágenes.
Se utiliza para cambiar las dimensiones de una imagen. En OpenCV, esta tarea se realiza con la función cv2.resize().

* Sintaxis: ``cv2.resize(src, dsize[, dst[, fx[, fy[, interpolation]]]])``

* Parámetros:

* ``src:`` imagen de entrada.

* ``dsize:`` tamaño deseado para la imagen de salida.

* ``fx y fy:`` factores de escala en los ejes horizontal y vertical, respectivamente.

* ``interpolation:`` método de interpolación.

Métodos de interpolación:

* ``cv2.INTER_LINEAR:`` buena opción predeterminada para agrandar o reducir imágenes.

* ``cv2.INTER_NEAREST:`` el más rápido, pero puede producir bordes irregulares.

* ``cv2.INTER_AREA:`` recomendado para reducir el tamaño, ya que evita el efecto muaré.

* ``cv2.INTER_CUBIC:`` más lento, pero da mejores resultados al ampliar imágenes.

* ``cv2.INTER_LANCZOS4:`` ofrece resultados de alta calidad.

````python
resized_image_1 = cv2.resize(image_rgb, (300, 300), interpolation=cv2.INTER_LINEAR)
print(f'image shape: {resized_image_1.shape}')
plt.imshow(resized_image_1)
plt.axis('off')
plt.show()
````

````python
resized_image_2 = cv2.resize(image_rgb, (0, 0), fx=0.5, fy=0.5)
print(f'image shape: {resized_image_2.shape}')
plt.imshow(resized_image_2)
plt.axis('off')
plt.show()
````
